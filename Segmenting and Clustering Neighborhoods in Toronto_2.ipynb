```python
import pandas as pd

url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

read_table = pd.read_html(url,header=[0])

df = read_table[0]

df = df[df.Borough !='Not assigned']

df.reset_index(drop=True,inplace=True)

df_grouped = df.groupby('Postal Code',as_index=False).agg(lambda x: ','.join(set(x.dropna())))

df_grouped.loc[df_grouped['Neighbourhood'] == 'Not assigned','Neighbourhood'] = df_grouped['Borough']

df_grouped.shape[0]
```

Now I will continue with the second part..

```python
# Read the  csv file that has the geographical coordinates of each postal code

df_codes = pd.read_csv('http://cocl.us/Geospatial_data')
df_codes.head()
df_codes.shape[0] # to check that we have the same number than the previous table
```

```python
# We need to cross the two tables

df_cross = pd.merge(left=df_grouped,right=df_codes,on='Postal Code')
df_cross.head()
```
